```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set random seed:
np.random.seed(123)
```

```python
# True parameter values
alpha, sigma = 1, 1
beta = [1, 2.5]

# Size of dataset
size = 100

# Predictor variable
X1 = np.linspace(0, 1, size)
X2 = np.linspace(0,.2, size)

# Simulate outcome variable
Y = alpha + beta[0]*X1 + beta[1]*X2 + np.random.randn(size)*sigma
```

```python
plt.scatter(X1,Y)
plt.show()
plt.scatter(X2,Y)
```
